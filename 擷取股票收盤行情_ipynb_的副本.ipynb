```python
import requests
import pandas as pd
import time
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
start_year = '2020'
start_month = '01'
end_year = '2021'
end_month = '09'
dates =[]
stockNo = 2330
url_template = "https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=html&date={}&stockNo={}"
data = pd.DataFrame()
```

```python
def append_dateslist(dates,yy,s_month,e_month):
    for mm in range(int(s_month),int(e_month)+1):
        dates.append('{}{}01'.format(str(yy),str(mm).zfill(2)))
    return dates
```

```python
for yy in range(int(start_year),int(end_year)+1):
  if int(start_year) == int(end_year):
      dates = append_dateslist(dates,yy,start_month,end_month) 
  elif yy == int(start_year):
      dates = append_dateslist(dates,yy,start_month,'12')
  elif yy == int(end_year):
      dates = append_dateslist(dates,yy,'01',end_month)
  else:
      dates = append_dateslist(dates,yy,'01','12')
```

```python
for date in dates :
    url = url_template.format(date, stockNo)
    read_data = pd.read_html(requests.get(url).text)[0]
    read_data.columns = read_data.columns.droplevel(0)
    data = pd.concat([data,read_data],ignore_index=True)
    print(data)
    time.sleep(0.1)
```

```python
file_name = "./drive/MyDrive/{}.csv".format(stockNo)
data.to_csv(file_name, index=False)
```
